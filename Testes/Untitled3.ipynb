```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Carregar os dados
data = pd.read_csv('winequality-white.csv', sep=';')

# Normalizar os dados (importante para PCA)
data_scaled = StandardScaler().fit_transform(data)

# Aplicar PCA para obter os trs primeiros componentes principais
pca = PCA(n_components=3)
pca_result = pca.fit_transform(data_scaled)

# Plotar os resultados em 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_result[:, 0], pca_result[:, 1], pca_result[:, 2], s=50, alpha=0.6, edgecolors='w', depthshade=True)

ax.set_xlabel('Componente 1')
ax.set_ylabel('Componente 2')
ax.set_zlabel('Componente 3')
ax.set_title('Visualizao 3D PCA de Wine Quality')
plt.show()
```
